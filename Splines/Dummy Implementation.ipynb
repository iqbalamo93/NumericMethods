```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as inter
```

```python
def f(x):
    return [0 if xx < 1 else 1 for xx in x]
```

Use `scipy.interpolate.CubicSpline` to construct an interpolating cubic spline and then plot the original function, the spline, and the knots used for interpolation.

Note that we don't know how to construct splines yet. Hence, we use the black box algorithm!

```python
np1 = 7         # number of knots
xmin = -1       # interval
xmax = 1

xk = np.linspace(xmin, xmax, np1)     #x input array
yk = f(xk)

X_inputs = np.linspace(xmin,xmax,100)   # create sampling points for plotting: x values
ykcs_nak = inter.CubicSpline(xk,yk,bc_type="not-a-knot")(xkplot) # cubic spline with not-a-knot boundaries
ykcs_nat = inter.CubicSpline(xk,yk,bc_type="natural")(xkplot)    # natural cubic spline
ykcs_cl = inter.CubicSpline(xk,yk,bc_type="clamped")(X_inputs)  # clamped cubic spline (zero first derivatives)
#ykcs_per = inter.CubicSpline(xk,yk,bc_type="periodic")(xkplot)   # periodic cubic spline (only for periodic f)
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.set_title("Cubic spline interpolation")
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.grid(linestyle="dotted")
ax.plot(X_inputs,ykcs_nak, label="not-a-knot")
ax.plot(X_inputs,ykcs_nat, label="natural")
ax.plot(X_inputs,ykcs_cl, label="clamped (zero first derivs)")
#ax.plot(xkplot,ykcs_per, label="periodic")
ax.plot(xk,yk,"o", label="knots")
ax.legend()

x_know = np.linspace(-1,9,num=100)
y_true =f(x_know)
ax.plot(x_know,y_true,"x", label="known")

plt.show()
```
