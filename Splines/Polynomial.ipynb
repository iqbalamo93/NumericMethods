```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as inter
```

Define the function which needs to be interpolated: $f(x)=\frac{1}{1+25(x+0.1)^2}$

```python
def f(x):
    return 1 / (1 + 25*(x+0.1)**2)
```

```python
order = 5                         # order of interpolating polynomial

x_true_points = np.linspace(-1, 1, order+1)  # create order+1 (x, y) knots used to construct
y_true_ponts = f(x_true_points)                        #  the interpolating polynomial

'''
IN real case we have only few true values of X and few true values of Y.
            EX: Yield cuve fitting
'''

x_input = np.linspace(-1, 1, 1000)#input
''' 
This is our input at which we want to predict Y
'''
y_true_sample  = f(x_true_sample)   #  IS UNKNOWN IN REAL                                    f
''' 
We don't know y_true_sample, our task is to guess it, here we know as we know our function.
we try to find answer hey! AT x what's the Y value.
'''

y_inter = inter.barycentric_interpolate(x_true_points, 
                                        y_true_ponts,
                                        x_input) # the polynomial evaluated at the 
'''
Here, we gave knw pairs of (x,y). And we guess value of unknown X
'''                                                        
```

```python
ax = plt.subplot(1, 1, 1) 
ax.plot(x_input, y_true_sample,  label="function_true")                         # plot original function
ax.plot(x_input, y_inter, label="polynomial_interpolation")                       # plot interpolating poly
ax.plot(x_true_points,     y_true_ponts, "o", label="knots") 
ax.set_title("Polynomial interpolation, order="+str(order))
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.annotate("$\\frac{1}{1+25(x+0.1)^2}$", size=20, xy=(-0.35,0.3), arrowprops=dict(arrowstyle="->"), xytext=(-0.5,0.0))
ax.grid(linestyle="dotted")
                           # plot knots xk, yk
ax.legend(loc="upper right")
plt.savefig("poly"+str(order)+".png", dpi=200)
plt.show()
```
